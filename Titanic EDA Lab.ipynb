```python
#Additional EDA for Age - distinguishing between adult and child passengers
#Distinction between adult and child based on Encyclopedia Titanica: https://www.encyclopedia-titanica.org/children-on-titanic/

age_bins = [0.0, 14.0, 80.0]
```

```python
age_names = ['child', 'adult']
```

```python
df.AgeCategory = pd.cut(df.Age, age_bins, labels=age_names)
```

```python
df.AgeCategory.value_counts(dropna=False)
```

```python
df.AgeCategory.value_counts(dropna=False).plot(kind='barh')
```

#### Age Summary:
<li>A continuous variable
<li>Value is missing for 177 of 891 records
<li>Ages range from 0.42 to 80
<li>May be useful to convert age data into a categorical variable, with categories of 'child' (14 or under) or 'adult'
<li>Further, may be useful to convert age data into a categorical variable with three categories: child, adult, senior
